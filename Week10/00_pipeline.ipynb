```python
%pip install -q papermill
import papermill as pm
from pathlib import Path
Path("mc/out").mkdir(parents=True, exist_ok=True)
```

```python
pm.execute_notebook(
    "fit_params.ipynb",
    "mc/out/_fit_params_run.ipynb",
    parameters=dict(
        TICKERS=["SPY","QQQ","GLD"],
        START="1999-01-01",
        END="2024-12-31",
        CSV="data/price_data.csv",    # or "" to use Yahoo
        OUT_DIR="mc"
    )
)
```

```python
pm.execute_notebook(
    "run_mc_experiments.ipynb",
    "mc/out/_mc_run.ipynb",
    parameters=dict(
        MODE="gaussian",           # "block" for block bootstrap
        DAYS=252*5,
        N_ITER=500,
        METHODS="momentum,sma_vix",
        COST_BPS="10,25,50",
        MGMT_FEES="0.01,0.02,0.03,0.04",
        PERF_FEES="0.05,0.10,0.20,0.25",
        VOL_MULT="1.0,1.5",
        SEED=2025,
        OUT_DIR="mc/out",
        PARAMS_PATH="mc/params.json",
        HIST_PATH="mc/hist_returns.csv"
    )
)
```
